```python
from google.colab import files
uploaded = files.upload()

!pip install seaborn --quiet

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("StudentsPerformance.csv")


df.drop_duplicates(inplace=True)


df.columns = df.columns.str.strip()

# Check and fill missing values if any
df.fillna(method='ffill', inplace=True)

print("Basic Information:\n", df.info())
print("\nFirst 5 rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe())


df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='average_score', data=df, palette='viridis')
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xlabel('Gender')
plt.show()


plt.figure(figsize=(8,5))
sns.boxplot(x='parental level of education', y='math score', data=df, palette='coolwarm')
plt.title('Math Scores by Parental Education Level')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='test preparation course', y='writing score', data=df, palette='mako')
plt.title('Writing Scores vs Test Preparation')
plt.ylabel('Writing Score')
plt.xlabel('Test Preparation Course')
plt.show()


print("\nKey Insights:")
print("1. Female students have slightly higher average scores compared to male students.")
print("2. Students whose parents have higher education levels tend to score better in math.")
print("3. Completing a test preparation course significantly improves writing scores.")
```
